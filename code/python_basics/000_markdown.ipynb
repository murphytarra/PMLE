## ðŸŽ› Dynamic Widgets: Control Parameters Easily

You can use sliders, dropdowns, and more to interactively adjust parameters in your analysis.

````python
import ipywidgets as widgets
from IPython.display import display

# Create sliders
slider = widgets.IntSlider(value=5, min=0, max=10, step=1, description="Slider:")
text_box = widgets.Text(value="Type here", description="Text Box:")

# Display widgets
display(slider, text_box)

# Use widget values dynamically
def update_slider(change):
    print(f"Slider value: {change['new']}")

slider.observe(update_slider, names='value')
````

## ðŸ—‚ Widgets for File Uploads

Let users upload files directly into the notebook.

````python
uploader = widgets.FileUpload()
display(uploader)

# Access the uploaded file
def on_upload_change(change):
    for filename, content in uploader.value.items():
        print(f"Uploaded file: {filename}")

uploader.observe(on_upload_change, names='value')
````
